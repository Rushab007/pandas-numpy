```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
data=pd.read_csv("trident.csv")
data
```

```python
data.set_index('Date',inplace=True)
```

```python
data
```

```python
plt.figure(figsize=(15,10))
plt.plot(data.index, data['Open'],color="g",linewidth=4.5,marker="o",markeredgecolor="r",markeredgewidth=5.5)
plt.xlabel("date")
plt.ylabel("price")
plt.title("TRIDENT Stock Price")
style.use("ggplot")
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.bar(data.index,data.Low,color="g",label="low prices")
plt.xlabel("date")
plt.ylabel("price")
plt.title("TRIDENT Stock Price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.scatter(data.index,data.Price,color="r")
plt.xlabel("date")
plt.ylabel("price")
plt.title("TRIDENT Stock Price")
style.use("dark_background")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(11,12))
plt.barh(data.index,data.Price,color="orange")
plt.xlabel("PRICE")
plt.ylabel("DATE")
plt.title("TRIDENT Stock Price")
plt.show()
```

```python
plt.boxplot(data.High)
plt.xlabel("date")
plt.ylabel("price")
plt.title("TRIDENT Stock Price")
plt.grid()
plt.show()
```
